```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
```

```python
michelin_restaurants_path = 'data/michelin_my_maps.csv'
yelp_restaurants_path = 'data/yelp_dataset/yelp_academic_dataset_business.json'
yelp_reviews_path = 'data/yelp_dataset/yelp_academic_dataset_review.json'
yelp_users_path = 'data/yelp_dataset/yelp_academic_dataset_user.json'
```

```python
michelin = pd.read_csv(michelin_restaurants_path)
yelp = pd.read_json(yelp_restaurants_path, lines=True)
```

```python
yelp_reviews = pd.read_json(yelp_reviews_path, lines=True)
```

```python
yelp_users = pd.read_json(yelp_users_path, lines=True)
```

```python
# Define columns that are allowed to have missing values and take only essentials
yelp_optional_cols = ['categories', 'hours', 'attributes']
michelin_optional_cols = ['Cuisine', 'PhoneNumber', 'Url', 'FacilitiesAndServices', 'Description', 'Price']
yelp_reviews_optional_cols = ['useful', 'funny', 'cool']
yelp_users_optional_cols = ['useful', 'funny', 'cool', 'fans', 'elite', 'friends', 'compliment_hot', 'compliment_more', 'compliment_profile', 'compliment_cute', 'compliment_list', 'compliment_note', 'compliment_plain', 'compliment_cool', 'compliment_funny', 'compliment_writer', 'compliment_photos']

# Essential
yelp_essential_cols = [col for col in yelp.columns if col not in yelp_optional_cols]
michelin_essential_cols = [col for col in michelin.columns if col not in michelin_optional_cols]
yelp_reviews_essential_cols = [col for col in yelp_reviews.columns if col not in yelp_reviews_optional_cols]
yelp_users_essential_cols = [col for col in yelp_users.columns if col not in yelp_users_optional_cols]

print(f"Yelp essential columns ({len(yelp_essential_cols)}): {yelp_essential_cols}")
print(f"Michelin essential columns ({len(michelin_essential_cols)}): {michelin_essential_cols}")
print(f"Yelp reviews essential columns ({len(yelp_reviews_essential_cols)}): {yelp_reviews_essential_cols}")
print(f"Yelp users essential columns ({len(yelp_users_essential_cols)}): {yelp_users_essential_cols}")
```

```python
# Create output directory
output_dir = 'data/cleaned'
os.makedirs(output_dir, exist_ok=True)
print(f"Created output directory: {output_dir}")
```

```python
# Remove rows with missing essential values from Yelp
yelp_has_missing_mask = yelp[yelp_essential_cols].isnull().any(axis=1)
print(f"Yelp rows with missing essential values: {yelp_has_missing_mask.sum()} ({yelp_has_missing_mask.sum()/len(yelp)*100:.2f}%)")
print(f"Yelp rows before cleaning: {len(yelp)}")

yelp_cleaned = yelp.loc[~yelp_has_missing_mask].copy()
print(f"Yelp rows after cleaning: {len(yelp_cleaned)}")
print(f"Yelp rows removed: {len(yelp) - len(yelp_cleaned)}")
```

```python
# Remove rows with missing essential values from Yelp Users
yelp_users_has_missing_mask = yelp_users[yelp_users_essential_cols].isnull().any(axis=1)
print(f"Yelp user rows with missing essential values: {yelp_users_has_missing_mask.sum()} ({yelp_users_has_missing_mask.sum()/len(yelp_users)*100:.2f}%)")
print(f"Yelp user rows before cleaning: {len(yelp_users)}")

yelp_users_cleaned = yelp_users.loc[~yelp_users_has_missing_mask].copy()
yelp_users_cleaned = yelp_users_cleaned.drop(columns=['...'], errors='ignore')
yelp_users_cleaned = yelp_users_cleaned.drop(columns=['friends'], errors='ignore')
print(f"Yelp user rows after cleaning: {len(yelp_users_cleaned)}")
print(f"Yelp user rows removed: {len(yelp_users) - len(yelp_users_cleaned)}")

# Randomly sample 50% of users
yelp_users_cleaned = yelp_users_cleaned.sample(frac=0.5, random_state=42)
print(f"Yelp user rows after 50% sampling: {len(yelp_users_cleaned)}")
```

```python
# Remove rows with missing essential values from Yelp Reviews
yelp_reviews_has_missing_mask = yelp_reviews[yelp_reviews_essential_cols].isnull().any(axis=1)
print(f"Yelp review rows with missing essential values: {yelp_reviews_has_missing_mask.sum()} ({yelp_reviews_has_missing_mask.sum()/len(yelp_reviews)*100:.2f}%)")
print(f"Yelp review rows before cleaning: {len(yelp_reviews)}")

yelp_reviews_cleaned = yelp_reviews.loc[~yelp_reviews_has_missing_mask].copy()
print(f"Yelp review rows after cleaning: {len(yelp_reviews_cleaned)}")
print(f"Yelp review rows removed: {len(yelp_reviews) - len(yelp_reviews_cleaned)}")

# Filter reviews to only include those from sampled users
sampled_user_ids = set(yelp_users_cleaned['user_id'])
yelp_reviews_cleaned = yelp_reviews_cleaned[yelp_reviews_cleaned['user_id'].isin(sampled_user_ids)]
print(f"Yelp review rows after filtering to sampled users: {len(yelp_reviews_cleaned)}")

# Randomly sample 50% of users
yelp_reviews_cleaned = yelp_reviews_cleaned.sample(frac=0.5, random_state=42)
print(f"Yelp review rows after 50% sampling: {len(yelp_users_cleaned)}")
```

```python
# Remove rows with missing essential values from Michelin
michelin_has_missing_mask = michelin[michelin_essential_cols].isnull().any(axis=1)
print(f"Michelin rows with missing essential values: {michelin_has_missing_mask.sum()} ({michelin_has_missing_mask.sum()/len(michelin)*100:.2f}%)")
print(f"Michelin rows before cleaning: {len(michelin)}")

michelin_cleaned = michelin.loc[~michelin_has_missing_mask].copy()
print(f"Michelin rows after cleaning: {len(michelin_cleaned)}")
print(f"Michelin rows removed: {len(michelin) - len(michelin_cleaned)}")
```

```python
# Export cleaned data
print("\n=== FINAL EXPORT SUMMARY ===")
print(f"Michelin final rows: {len(michelin_cleaned):,}")
print(f"Yelp business final rows: {len(yelp_cleaned):,}")
print(f"Yelp reviews final rows: {len(yelp_reviews_cleaned):,}")
print(f"Yelp users final rows: {len(yelp_users_cleaned):,}")
```

```python
# Export Yelp data
yelp_output_path = os.path.join(output_dir, 'yelp_cleaned.csv')
yelp_cleaned.to_csv(yelp_output_path, index=False)
print(f"✓ Exported Yelp data to: {yelp_output_path}")
print(f"  - Rows: {len(yelp_cleaned):,}")
print(f"  - Size: {os.path.getsize(yelp_output_path) / 1024:.2f} KB")
```

```python
# Export Yelp reviews data
yelp_review_output_path = os.path.join(output_dir, 'yelp_reviews_cleaned.csv')
yelp_reviews_cleaned.to_csv(yelp_review_output_path, index=False)
print(f"✓ Exported Yelp reviews to: {yelp_review_output_path}")
print(f"  - Rows: {len(yelp_reviews_cleaned):,}")
print(f"  - Size: {os.path.getsize(yelp_review_output_path) / 1024:.2f} KB")
```
