```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
```

```python
michelin_restaurants_path = 'data/michelin_my_maps.csv'
yelp_restaurants_path = 'data/yelp_dataset/yelp_academic_dataset_business.json'
yelp_reviews_path = 'data/yelp_dataset/yelp_academic_dataset_review.json'
```

```python
michelin = pd.read_csv(michelin_restaurants_path)
yelp = pd.read_json(yelp_restaurants_path, lines=True)
yelp_reviews = pd.read_json(yelp_reviews_path, lines=True)
```

```python
# Define columns that are allowed to have missing values and take only essentials
yelp_optional_cols = ['categories', 'hours', 'attributes']
michelin_optional_cols = ['Cuisine', 'PhoneNumber', 'Url', 'FacilitiesAndServices', 'Description', 'Price']
yelp_reviews_optional_cols = ['useful', 'funny', 'cool']

# Essential
yelp_essential_cols = [col for col in yelp.columns if col not in yelp_optional_cols]
michelin_essential_cols = [col for col in michelin.columns if col not in michelin_optional_cols]
yelp_reviews_essential_cols = [col for col in yelp_reviews.columns if col not in yelp_reviews_optional_cols]

print(f"Yelp essential columns ({len(yelp_essential_cols)}): {yelp_essential_cols}")
print(f"Michelin essential columns ({len(michelin_essential_cols)}): {michelin_essential_cols}")
print(f"Yelp reviews essential columns ({len(yelp_reviews_essential_cols)}): {yelp_reviews_essential_cols}")
```

```python
# Create output directory
output_dir = 'data/cleaned'
os.makedirs(output_dir, exist_ok=True)
print(f"Created output directory: {output_dir}")
```

```python
# Remove rows with missing essential values from Yelp
yelp_has_missing_mask = yelp[yelp_essential_cols].isnull().any(axis=1)
print(f"Yelp rows with missing essential values: {yelp_has_missing_mask.sum()} ({yelp_has_missing_mask.sum()/len(yelp)*100:.2f}%)")
print(f"Yelp rows before cleaning: {len(yelp)}")

yelp_cleaned = yelp.loc[~yelp_has_missing_mask].copy()
print(f"Yelp rows after cleaning: {len(yelp_cleaned)}")
print(f"Yelp rows removed: {len(yelp) - len(yelp_cleaned)}")
```

```python
# Remove rows with missing essential values from Yelp Reviews
yelp_reviews_has_missing_mask = yelp_reviews[yelp_reviews_essential_cols].isnull().any(axis=1)
print(f"Yelp review rows with missing essential values: {yelp_reviews_has_missing_mask.sum()} ({yelp_reviews_has_missing_mask.sum()/len(yelp_reviews)*100:.2f}%)")
print(f"Yelp review rows before cleaning: {len(yelp_reviews)}")

yelp_reviews_cleaned = yelp_reviews.loc[~yelp_reviews_has_missing_mask].copy()
print(f"Yelp review rows after cleaning: {len(yelp_reviews_cleaned)}")
print(f"Yelp review rows removed: {len(yelp_reviews) - len(yelp_reviews_cleaned)}")
```

```python
# Remove rows with missing essential values from Michelin
michelin_has_missing_mask = michelin[michelin_essential_cols].isnull().any(axis=1)
print(f"Michelin rows with missing essential values: {michelin_has_missing_mask.sum()} ({michelin_has_missing_mask.sum()/len(michelin)*100:.2f}%)")
print(f"Michelin rows before cleaning: {len(michelin)}")

michelin_cleaned = michelin.loc[~michelin_has_missing_mask].copy()
print(f"Michelin rows after cleaning: {len(michelin_cleaned)}")
print(f"Michelin rows removed: {len(michelin) - len(michelin_cleaned)}")
```

```python
# Analyze Yelp country distribution
US_STATES = {'AL','AK','AZ','AR','CA','CO','CT','DC','DE','FL','GA','HI','IA','ID','IL','IN','KS','KY','LA','MA','MD','ME','MI','MN','MO','MS','MT','NC','ND','NE','NH','NJ','NM','NV','NY','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VA','VT','WA','WI','WV','WY'}
CA_PROVINCES = {'AB','BC','MB','NB','NL','NS','NT','NU','ON','PE','QC','SK','YT'}

def classify_yelp_country(row):
    state = str(row.get('state', '')).strip().upper()
    postal = str(row.get('postal_code', '')).strip()
    if state in US_STATES or (postal.isdigit() and len(postal) == 5):
        return 'United States'
    if state in CA_PROVINCES:
        return 'Canada'
    return 'Other/Unknown'

yelp['country_classified'] = yelp.apply(classify_yelp_country, axis=1)
country_counts = yelp['country_classified'].value_counts()

print("Yelp FULL dataset country distribution:")
print(country_counts)
print(f"\nTotal: {len(yelp):,}")
print(f"US: {country_counts.get('United States', 0):,} ({country_counts.get('United States', 0)/len(yelp)*100:.2f}%)")
print(f"Canada: {country_counts.get('Canada', 0):,} ({country_counts.get('Canada', 0)/len(yelp)*100:.2f}%)")
print(f"Other/Unknown: {country_counts.get('Other/Unknown', 0):,} ({country_counts.get('Other/Unknown', 0)/len(yelp)*100:.2f}%)")

yelp = yelp.drop(columns=['country_classified'])
```

```python
# Match restaurants using name similarity and geographic proximity

def standardize_text(text):
    """Standardize text for matching: lowercase, strip whitespace"""
    if pd.isna(text):
        return ''
    return str(text).lower().strip()

# Standardize names for both datasets
michelin_cleaned['name_std'] = michelin_cleaned['Name'].apply(standardize_text)
yelp_cleaned['name_std'] = yelp_cleaned['name'].apply(standardize_text)

# Calculate distance between two lat/lon points in kilometers
def distance(lat1, lon1, lat2, lon2):
    """Calculate distance between two points on Earth in kilometers"""
    from math import radians, sin, cos, sqrt, atan2
    
    R = 6371
    
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    return R * c

# Match threshold: restaurants within 16 km (5 miles) with same name
distance_threshold_km = 16.0

matches = []
print(f"Michelin restaurants: {len(michelin_cleaned)}")
print(f"Yelp businesses: {len(yelp_cleaned)}")
print(f"Matching with distance threshold: {distance_threshold_km} km\n")

# For each Michelin restaurant, find matching Yelp business
for idx, mich_row in tqdm(michelin_cleaned.iterrows(), total=len(michelin_cleaned), desc="Matching restaurants"):
    mich_name = mich_row['name_std']
    mich_lat = mich_row['Latitude']
    mich_lon = mich_row['Longitude']
    
    # Find Yelp businesses with same name
    yelp_same_name = yelp_cleaned[yelp_cleaned['name_std'] == mich_name]
    
    # Check distance
    for yelp_idx, yelp_row in yelp_same_name.iterrows():
        yelp_lat = yelp_row['latitude']
        yelp_lon = yelp_row['longitude']
        
        distance = distance(mich_lat, mich_lon, yelp_lat, yelp_lon)
        
        if distance <= distance_threshold_km:
            matches.append({
                'michelin_idx': idx,
                'yelp_idx': yelp_idx,
                'name': mich_name,
                'distance_km': distance
            })

print(f"\nCommon restaurants found: {len(matches)}")

# Extract matched indices
if len(matches) > 0:
    michelin_matched_indices = [m['michelin_idx'] for m in matches]
    yelp_matched_indices = [m['yelp_idx'] for m in matches]
    
    # Filter to only matched restaurants
    michelin_cleaned = michelin_cleaned.loc[michelin_matched_indices].copy()
    yelp_cleaned = yelp_cleaned.loc[yelp_matched_indices].copy()
    
    print(f"Michelin restaurants after matching: {len(michelin_cleaned)}")
    print(f"Yelp businesses after matching: {len(yelp_cleaned)}")
    
    
    print(f"\nSample matches:")
    for match in matches[:5]:
        print(f"  {match['name']} - distance: {match['distance_km']*1000:.1f} meters")
else:
    print("No matches found!")
```

```python
# Prepare data for export (drop temporary columns if present)
columns_to_drop = ['name_std']
michelin_export = michelin_cleaned.drop(columns=[c for c in columns_to_drop if c in michelin_cleaned.columns])
yelp_export = yelp_cleaned.drop(columns=[c for c in columns_to_drop if c in yelp_cleaned.columns])
# Reviews have no temp columns
yelp_review_export = yelp_reviews_cleaned.copy()
print(f"Prepared data for export: Michelin={michelin_export.shape}, Yelp Businesses={yelp_export.shape}, Yelp Reviews={yelp_review_export.shape}")
print("\n=== FINAL EXPORT SUMMARY ===")
print(f"Michelin final rows: {len(michelin_export):,}")
print(f"Yelp business final rows: {len(yelp_export):,}")
print(f"Yelp reviews final rows: {len(yelp_review_export):,}")
```

```python
# Export Yelp data
yelp_output_path = os.path.join(output_dir, 'yelp_cleaned.csv')
yelp_export.to_csv(yelp_output_path, index=False)
print(f"✓ Exported Yelp data to: {yelp_output_path}")
print(f"  - Rows: {len(yelp_export):,}")
print(f"  - Size: {os.path.getsize(yelp_output_path) / 1024:.2f} KB")
```

```python
# Export Yelp reviews data
yelp_review_output_path = os.path.join(output_dir, 'yelp_reviews_cleaned.csv')
yelp_review_export.to_csv(yelp_review_output_path, index=False)
print(f"✓ Exported Yelp reviews to: {yelp_review_output_path}")
print(f"  - Rows: {len(yelp_review_export):,}")
print(f"  - Size: {os.path.getsize(yelp_review_output_path) / 1024:.2f} KB")
```
