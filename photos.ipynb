```python
import json
```

```python
input_file = "/Users/grace/Downloads/Yelp/yelp_photos/photos.json"          # your existing file
output_file = "/Users/grace/Downloads/Yelp/yelp_photos/photos_with_urls.json"  # new file to write
bucket = "dishcord-yelp-photos"
prefix = "uploaded-images/./"        # your exact key prefix
extension = ".jpg"
```

```python
def build_url(photo_id):
    key = f"{prefix}{photo_id}{extension}"
    # Return proper S3 URI format
    return f"s3://{bucket}/{key}"

new_rows = []

with open(input_file, "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        
        obj = json.loads(line)
        photo_id = obj["photo_id"]

        # Add aws_url to the object
        obj["aws_url"] = build_url(photo_id)

        new_rows.append(obj)

# Write out a new JSON lines file
with open(output_file, "w") as f:
    for row in new_rows:
        f.write(json.dumps(row) + "\n")
```
